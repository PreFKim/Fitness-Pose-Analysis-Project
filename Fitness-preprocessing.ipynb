```python
palette = [(255, 165, 0),(0, 128, 0),(128, 0, 128),(255, 0, 255),(0, 0, 128),(128, 128, 0),(0, 0, 255),(165, 42, 42),(0, 255, 255),(255, 192, 203),(70, 130, 90),(255, 99, 71),(0, 128, 128),(255, 0, 0),(0, 255, 0),(0, 0, 0),(255, 255, 255),(128, 128, 128),(210, 105, 30),(218, 165, 32),(173, 216, 230),(255, 20, 147),(0, 255, 127),(220, 20, 60)]
classes = ["프런트 레이즈",
           "업라이트로우",
           "바벨 스티프 데드리프트",
           "바벨 로우",
           "덤벨 벤트오버 로우",
           "바벨 데드리프트",
           "바벨 스쿼트",
           "오버 헤드 프레스",
           "사이드 레터럴 레이즈",
           "바벨 컬 ",
           "덤벨 컬",
           "덤벨 풀 오버"]

point_name = ['Nose', 'Left Eye', 'Right Eye', 'Left Ear', 'Right Ear', 'Left Shoulder', 'Right Shoulder', 'Left Elbow', 'Right Elbow', 'Left Wrist', 'Right Wrist', 'Left Hip', 'Right Hip', 'Left Knee', 'Right Knee', 'Left Ankle', 'Right Ankle', 'Neck', 'Left Palm', 'Right Palm', 'Back', 'Waist', 'Left Foot', 'Right Foot']
linked_name =  [["Left Ear","Left Eye"],["Left Eye","Nose"],["Right Ear","Right Eye"],["Right Eye","Nose"],["Nose","Neck"],["Left Palm","Left Wrist"],["Left Wrist","Left Elbow"],["Left Elbow","Left Shoulder"],["Left Shoulder","Neck"],["Right Palm","Right Wrist"],["Right Wrist","Right Elbow"],["Right Elbow","Right Shoulder"],["Right Shoulder","Neck"],["Neck","Back"],["Back","Waist"],["Left Foot","Left Ankle"],["Left Ankle","Left Knee"],["Left Knee","Left Hip"],["Left Hip","Waist"],["Right Foot","Right Ankle"],["Right Ankle","Right Knee"],["Right Knee","Right Hip"],["Right Hip","Waist"]]

for i in range(len(linked_name)):
    linked_name[i][0] = point_name.index(linked_name[i][0])
    linked_name[i][1] = point_name.index(linked_name[i][1])
```

```python


def draw_joint(img,pts,rad=3,line=1):
    out = img.copy()
    h,w,c = img.shape
    used = np.zeros_like(point_name,dtype=np.int32())
    #print(used)
    if np.max(pts) <=1:
        pts = pts * np.array([w,h])

    for i,se in enumerate(linked_name):
        from_xy = (int(pts[se[0]][0]),int(pts[se[0]][1]))
        to_xy = (int(pts[se[1]][0]),int(pts[se[1]][1]))

        out = cv2.line(out,from_xy,to_xy,[255,255,255],line,cv2.LINE_AA)
        if (used[se[0]]==0):
            out= cv2.circle(out,from_xy,rad,palette[se[0]],-1,cv2.LINE_AA)
            used[se[0]] = 1

        if (used[se[1]]==0):
            out= cv2.circle(out,to_xy,rad,palette[se[1]],-1,cv2.LINE_AA)
            used[se[1]] = 1

    return out
```

```python
dataset = [[],[],[],[],[]]

worker = 64
def load_file(filename):
    with open(filename) as f:
        label_json = json.load(f)
        for view in range(5):
            if len(label_json['frames']) != 0 :
                path_parts = filename.split("/")
                cropped_path = ("/".join(path_parts[:filename.count('/')-1])).replace("label","video")
                minx=1920
                miny=1080
                maxx=0
                maxy=0

                tmp = []
                video_path = os.path.dirname(cropped_path + '/' + label_json['frames'][0][f'view{view+1}']['img_key'])+'.avi'
                for frame in range(len(label_json['frames'])):
                    pts_list = []

                    # 좌표 추가
                    for p in point_name:
                        pts_list.append([label_json['frames'][frame][f'view{view+1}']['pts'][p]['x'],
                                         label_json['frames'][frame][f'view{view+1}']['pts'][p]['y']])
                        minx = min(minx,pts_list[-1][0])
                        miny = min(miny,pts_list[-1][1])
                        maxx = max(maxx,pts_list[-1][0])
                        maxy = max(maxy,pts_list[-1][1])


                    tmp.append([video_path,frame,[],pts_list,label_json['type_info']['exercise']])

                for frame in range(len(label_json['frames'])):
                    tmp[frame][2]=[minx,miny,maxx,maxy]
                dataset[view].append(tmp)



start = 0
for i in tqdm.tqdm(range(worker*start,len(file_list[0:worker*(start+2)]),worker)):
    with concurrent.futures.ThreadPoolExecutor(max_workers=worker) as executor:
        executor.map(load_file, file_list[i:i+worker])

for i in range(1,5):

    if len(dataset[i]) != len(dataset[i-1]) :
        print(i, "error")

dataset = [sorted(lst) for lst in dataset]

print()
print(dataset[0][0][0]) #view,data,frame,value
print(dataset[0][1][0])
print(dataset[0][2][0])
print(dataset[0][3][0])
```

```python
def copy_tree(lst):
    filename = lst[0][0]
    from_dir = os.path.dirname(filename)+'/'
    to_dir = from_dir.replace(root_path,copy_path)
    if os.path.exists(to_dir)==False:
        os.makedirs(to_dir,exist_ok=True)
    shutil.copy(filename,filename.replace(root_path,copy_path))

for i in range(5):
    for j in tqdm.tqdm(range(0,len(dataset[i]),worker)):
        with concurrent.futures.ThreadPoolExecutor(max_workers=worker) as executor:
            executor.map(copy_tree, dataset[i][j:j+worker])
```

```python


def crop(day,from_to=[],del_dir=True,training=True):
    count=0
    if training:
        root_path = '/content/drive/MyDrive/fitness/dataset/training'
    else:
        root_path = '/content/drive/MyDrive/fitness/dataset/validation'
    original_size = 0.0
    croped_size = 0.0
    error = []
    label_num = 0
    for labelname in sorted(glob.glob(f'{root_path}/label/Day{day:02d}*/*[!d].json')):
        if label_num != int(labelname.split('-')[-2]):
            label_num = int(labelname.split('-')[-2])
            print(label_num,"탐색 시작")

        with open(labelname, "r") as st_json:
            label = json.load(st_json)

        if (len(label['frames'])!=0):

            #사람만을 crop하기 위한
            for num in range(1,6):
                minx=1920
                miny=1080
                maxx=0
                maxy=0
                view = "view"+str(num)
                for i in range(len(label['frames'])):
                    for key in label['frames'][i][view]['pts'].keys():
                        minx = min(minx,label['frames'][i][view]['pts'][key]['x'])
                        maxx = max(maxx,label['frames'][i][view]['pts'][key]['x'])
                        miny = min(miny,label['frames'][i][view]['pts'][key]['y'])
                        maxy = max(maxy,label['frames'][i][view]['pts'][key]['y'])

                img_save_dir = os.path.dirname(root_path+'/img/'+label['frames'][0][view]['img_key'])

                os.makedirs(img_save_dir,exist_ok=True)

                if (len(label['frames'])!=len(glob.glob(img_save_dir+'/*.jpg'))):

                    #라벨 내에 있는 프레임 별 모든 데이터

                    for i in range(len(label['frames'])):
                        read_dir = '/content/'+label['frames'][i][view]['img_key']
                        write_dir = root_path+'/img/'+label['frames'][i][view]['img_key']

                        if (os.path.exists(write_dir) == False):
                            if (os.path.exists(read_dir) == False):
                                if len(from_to)==0:
                                    error.append(read_dir)
                                    print("개수가0")
                                    continue
                                else:
                                    find = 0
                                    for ft in from_to:
                                        from_dir = read_dir.replace(ft[0],ft[1])
                                        if (os.path.exists(from_dir)):
                                            read_dir = from_dir
                                            find = 1
                                            break
                                    if find != 1:
                                        error.append(read_dir)
                                        continue

                        img = cv2.imread(read_dir)

                        if type(img)==type(None):
                            print(read_dir)

                        #print(img.shape)# 1080,1920,3
                        crop_size = 200

                        crop_min_y = max(0,miny-crop_size)
                        crop_max_y = min(img.shape[0],maxy+crop_size)
                        crop_min_x = max(0,minx-crop_size)
                        crop_max_x = min(img.shape[1],maxx+crop_size)
                        crop = img[crop_min_y:crop_max_y,crop_min_x:crop_max_x,:]

                        cv2.imwrite(write_dir,crop,[cv2.IMWRITE_JPEG_QUALITY, 50])

                        #print(label['type_info']['type'],label['frames'][i][view]['img_key'],os.path.getsize(read_dir)//1024,os.path.getsize(write_dir)//1024,img.shape,crop.shape)

                        original_size += os.path.getsize(read_dir)/1024
                        croped_size += os.path.getsize(write_dir)/1024
                        count+=1
    print("기존 파일 용량",original_size,"Kb")
    print("결과 파일 용량",croped_size,"Kb")
    print("개수",count)
    if (croped_size!=0):
        print("용량 비율",original_size/croped_size)
    print("error",error)

    if(del_dir):
        delete = glob.glob(f"/content/Day{day:02d}*")[-1]
        shutil.rmtree(delete)
```
